```python
my_list = [1, 2, 3, 4, 5]
print(my_list[0])  # Output: 1
print(my_list[2])  # Output: 3
print(my_list[-1]) # Output: 5
```

```python
# List methods
my_list.append(6)  # Adds 6 to the end of the list
print(my_list)  # Output: [1, 2, 3, 4, 5, 6]
my_list.insert(0, 0)  # Inserts 0 at index 0
print(my_list)  # Output: [0, 1, 2, 3, 4, 5, 6]
my_list.remove(3)  # Removes the first occurrence of 3
print(my_list)  # Output: [0, 1, 2, 4, 5, 6]
my_list.pop()  # Removes and returns the last item
print(my_list)  # Output: [0, 1, 2, 4, 5]
my_list.reverse()  # Reverses the list
print(my_list)  # Output: [5, 4, 2, 1, 0]
my_list.sort()  # Sorts the list in ascending order
print(my_list)  # Output: [0, 1, 2, 4, 5]   
my_list.sort(reverse=True)  # Sorts the list in descending order
print(my_list)  # Output: [5, 4, 2, 1, 0]
my_list.clear()  # Removes all items from the list
print(my_list)  # Output: []
```

```python
#create a list containing table of 5

table_of_6 = [6 * i for i in range(1, 11)]
print(table_of_6)  # Output: [5, 10, 15, 20, 25, 30, 35, 40, 45, 50
```

```python
squared = [i**2 for i in range(5)]
print(squared)  # Output: [0, 1, 4, 9, 16]
```

```python
# Tuples and operations on tuples
my_tuple = (1, 2, 3, 4, 5)
print(my_tuple[0])  # Output: 1
print(my_tuple[2])  # Output: 3
print(my_tuple[-1]) # Output: 5
# Tuples are immutable, so we cannot modify them directly
# However, we can create a new tuple by concatenating existing tuples
new_tuple = my_tuple + (6, 7)
print(new_tuple)  # Output: (1, 2, 3, 4, 5, 6, 7)
# We can also use tuple unpacking to assign values to variables
a, b, c,d,e = my_tuple
print(a, b, c,d,e)  # Output: 1 2
print(my_tuple)  # Output: 1
print(my_tuple.count(2))  # Output: 1
print(my_tuple.index(3))  # Output: 2   
```
