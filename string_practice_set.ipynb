```python
'''
Basic String Operations in python
1. Create a string variable name with your full name. Print;
    1. The First character of the string
    2. The last character of the string
    3. The length of the string

2. Concatenate two string: "Hello" and "World". with a space in between.
'''

name = "John Doe"
print(name[0]) # prints the first character of the string
print(name[-1]) # prints the last character of the string
print(len(name)) # prints the length of the string

str1 = "Hello"
str2 = "World"
result = str1 + " " + str2 # concatenates the two strings with a space
print(result) # prints "Hello World"
```

```python
'''
1. String Slicing and indexing
    1. Given the string "Python Programming", do the following:
        - print the first 6 characters
        - print the last 6 characters
        - print every second character from the string
    2. Given the string "Data Science", extract the word "Data" using slicing.
    3. Given the string "Machine Learning", extract the word "Learning" using slicing.
2. Reverse the string 'text' using slicing.
'''

string = "Python Programming"
print(string[:6]) # prints the first 6 characters
print(string[-6:]) # prints the last 6 characters
print(string[::2]) # prints every second character from the string

s2 = 'text'
reversed_s2 = string[::-1] # reverses the string using slicing
print(reversed_s2) # prints 'txet'
```

```python
'''
1. String Methods and Functions
    1. Take the string " i love python programming " and do the following:
        - Remove extra spaces from both ends
        - Convert it to title case
        - Count how many times the letter 'o' appears in the string
    2. Check if the string "123abc" is alphanumeric.
'''
s3 = " i love python programming "
s3 = s3.strip() # removes extra spaces from both ends
s3 = s3.title() # converts the string to title case
print(s3) # prints "I Love Python Programming"
count_o = s3.count('o') # counts how many times the letter 'o'
print(count_o) # prints the count of 'o'
s4 = "123abc"
is_alphanumeric = s4.isalnum() # checks if the string is alphanumeric
print(is_alphanumeric) # prints True
```

```python
'''
1. String Formatting and f-strings
    1. using format(), create a sentence
      "My name is john and I am 25 years old."
      by passing "john" and "25" as variables.
    2. using f-strings, create the same sentence as above.
'''
name = "john"
age = 25
sentence = "My name is {} and I am {} years old.".format(name, age)
print(sentence) # prints "My name is john and I am 25 years old."
sentence_fstring = f"My name is {name} and I am {age} years old."
print(sentence_fstring) # prints "My name is john and I am 25 years old."
```

```python
'''
1. String Manipulation Challenges
    1. Given sentence = "Coding in Python is fun", replace "fun" with "awesome" and print it.
    2. Find the index of the word "Python" in sentence.capitalize
    3. Convert the entire sentence to uppercase and print it.
'''
sentence = "Coding in Python is fun"
sentence = sentence.replace("fun", "awesome") # replaces "fun" with "awesome"
print(sentence) # prints "Coding in Python is awesome"
index_python = sentence.find("Python") # finds the index of the word "Python"
print(index_python) # prints the index of "Python"`
sentence_upper = sentence.upper() # converts the entire sentence to uppercase
print(sentence_upper) # prints "CODING IN PYTHON IS AWESOME"
```
