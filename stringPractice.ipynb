```python
'''string methods'''
name = 'Pankaj saxena'
a = len(name) #length of string
print(a) #length of string
b = name.upper() #converts to uppercase
print(b) #prints the string in uppercase
c = name.lower() #converts to lowercase
print(c) #prints the string in lowercase
d = name.title() #converts to title case
print(d) #prints the string in title case
e = name.count('a') #counts the number of occurrences of 'a'
print(e) #prints the count of 'a'
f = name.find('s') #finds the index of first occurrence of 's'
print(f) #prints the index of first occurrence of 's'
g = name.replace('a', 'o') #replaces 'a' with 'o
print(g) #prints the string with 'a' replaced by 'o'
```

```python
# string split and joining example
sentence = "Hello World"
words = sentence.split() # splits the sentence into a list of words
print(words) # prints the list of words
joined_sentence = ' '.join(words) # joins the list of words back into a sentence
print(joined_sentence) # prints the joined sentence 

text = "apples, oranges, bananas"
fruits = text.split(', ') # splits the text into a list of fruits
print(fruits) # prints the list of fruits
joined_fruits = ' - '.join(fruits) # joins the list of fruits with ' and '
print(joined_fruits) # prints the joined fruits
```

```python
# string formatting and f-strings example
name = "Alice"
age = 30
# using string formatting
formatted_string = "My name is {} and I am {} years old.".format(name, age)
print(formatted_string) # prints the formatted string   
# using f-strings
f_string = f"My name is {name} and I am {age} years old."
print(f_string) # prints the f-string
```

```python
# ord() and chr() functions example
char = 'A'
ascii_value = ord(char) # gets the ASCII value of the character
print(ascii_value) # prints the ASCII value of 'A'
ascii_value = 66
char = chr(ascii_value) # gets the character corresponding to the ASCII value
print(char) # prints the character corresponding to ASCII value 66, which is 'B'
```

```python
'''
Input: 
a = Hi, b = There
Output: 
HiThereHi
Explanation: 
After joining short+long+short strings, 
we have new string as "HiThereHi".
'''

a = "Pankaj Saxena"
b = "There"
if len(a) < len(b):
    result = a + b + a
else:
    result = b + a + b
print(result) # prints the result of joining short+long+short strings
```
