```python
#docstring example
def add(a, b):
    """
    This function takes two numbers and returns their sum.
    
    Parameters:
    a (int): The first number.
    b (int): The second number.
    
    Returns:
    int: The sum of a and b.
    """
    return a + b

print(add.__doc__)
print(add(3, 5))
```
