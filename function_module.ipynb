```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))
print(greet("Bob"))

def avarage_find(a,b,c):
    return (a + b + c) / 3

print(avarage_find(10, 20, 30))
print(avarage_find(5, 15, 25))

```

```python
# Function arguments & Return Values
def square(x):
    return x * x    

print(square(5))
print(square(10))
```

```python
# Lambda functions
square_lambda = lambda x: x * x
print(square_lambda(5))
print(square_lambda(10))

numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x * x, numbers))
print(squared_numbers)

sum = lambda a, b: a + b
print(sum(5, 10))
```

```python
# Recursion in Python
# A function calling itself to solve a problem
# Example for factorial program

def factorial(n):
    if(n==1):
        return 1
    
    return n * factorial(n-1)

print(factorial(5)) #120

```

```python
#febonaci series

def fibonacci(n):
    if(n == 0 or n==1):
        return n
    return fibonacci(n-2) + fibonacci(n-1)

number = 10
for i in range(number):
    print(fibonacci(i), end=" ") 
```

```python
#4. Generate Fibonacci List
def feb_list(n):
    fib = [0,1]
    for i in range(2,n):
        fib.append(fib[i-1] + fib[i-2])
    return fib[:n]

print(feb_list(12))
```
