                                                                                  Function and Modules

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))
print(greet("Bob"))

def avarage_find(a,b,c):
    return (a + b + c) / 3

print(avarage_find(10, 20, 30))
print(avarage_find(5, 15, 25))

```

```python
# Function arguments & Return Values
def square(x):
    return x * x    

print(square(5))
print(square(10))
```

```python
# Lambda functions
square_lambda = lambda x: x * x
print(square_lambda(5))
print(square_lambda(10))

numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x * x, numbers))
print(squared_numbers)

sum = lambda a, b: a + b
print(sum(5, 10))
```
